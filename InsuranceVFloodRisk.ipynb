```python
import pandas as pd
import numpy as np
from scipy.stats import ttest_ind
from scipy.stats import pearsonr
import statsmodels.formula.api as smf
```

```python
nri = pd.read_csv('NRI_Table_Counties.csv')
needed = []
for i in range(len(list(nri['STATE']))):
  needed.append([list(nri['STATE'])[i],list(nri['COUNTY'])[i],list(nri['RFLD_RISKS'])[i]])

needed = list(needed)
```

```python
insurance = pd.read_csv('Insurance cost for states and counties.csv')
insurancemat = insurance.to_numpy()
insurancematuse = list(insurancemat)
for i in range(len(insurancematuse)):
  insurancematuse[i] = list(insurancematuse[i])

states = []
for i in range(len(insurancematuse)):
  states.append(insurancematuse[i][2])

states = sorted(list(set(states)))

eachstate = []

for l in range(len(states)):
  for i in range(len(needed)):
    if needed[i][0] == states[l]:
      eachstate.append(needed[i][2])

for i in range(len(insurancematuse)):
  if len(insurancematuse[i][1]) == 5:
    insurancematuse[i][1] = insurancematuse[i][1][:1] + insurancematuse[i][1][2:]
  insurancematuse[i][1] = int(insurancematuse[i][1])
  insurancematuse[i].append(eachstate[i])
  insurancematuse[i] = insurancematuse[i]
```

```python
insuranceflood = pd.DataFrame(insurancematuse)
insuranceflood.columns = ["County","Cost of insurance","State","Flood Risk Index"]
insuranceflood
```

```python
cost = list(insuranceflood["Cost of insurance"])
risk = list(insuranceflood["Flood Risk Index"])

info = pearsonr(cost, risk)

print("Correlation coefficent: "+ str(info[0]))
print("p-value: "+ str(info[1]))
```

```python
newdf = []

for i in range(len(cost)):
  newdf.append([cost[i],risk[i]])

newdf = pd.DataFrame(newdf)

model = smf.ols('risk ~ cost', data=newdf)
model = model.fit()
model.params
```
